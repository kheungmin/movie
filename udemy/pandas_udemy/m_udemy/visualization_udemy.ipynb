```python
# 히트맵 이용하여 상관관계 확인하기
sns.heatmap(df.corr(), annot=True, linewidths=1)
plt.show()
```

- 값이 -1.0 ~ -0.7 이면, 강한 음적 상관관계
- 값이 -0.7 ~ -0.3 이면, 뚜렷한 음적 상관관계
- 값이 -0.3 ~ -0.1 이면, 약한 음적 상관관계
- 값이 -0.1 ~ +0.1 이면, 없다고 할 수 있는 상관관계
- 값이 +0.1 ~ +0.3 이면, 약한 양적 상관관계
- 값이 +0.3 ~ +0.7 이면, 뚜렷한 양적 상관관계
- 값이 +0.7 ~ +1.0 이면, 강한 양적 상관관계

```python
df_temp=df.copy()
df_temp.head()
```

```python
temp = df_temp.drop(['category'], axis=1)
temp_sort = temp.sort_values(by='num_students',ascending=False).iloc[:10]

plt.barh(temp_sort['title'], temp_sort['num_students'], align='center', alpha=0.8, color='green')
plt.ylabel('강의이름')
plt.title("수강생 기준 상위 10개 강좌")
plt.show()
```

```python
temp_sort.sort_values(by='num_students', ascending=True)
```

```python
temp = df_temp.drop(['category'], axis=1)
temp_sort = temp.sort_values(by='num_students',ascending=False).iloc[:10]

temp_sort_desc = temp_sort.sort_values(by='num_students')
plt.barh(temp_sort_desc['title'], temp_sort_desc['num_students'], align='center', alpha=0.8, color='green')
plt.ylabel('강의이름')
plt.title("수강생 기준 상위 10개 강좌")
plt.show()
```

```python
df_sorted_num_student=df_temp.sort_values(by='num_students', ascending=False).drop_duplicates(['title','num_videos','name_instructor','category','stars','num_students','total_time'])
df_sorted_num_student
```

```python
df_sorted_num_student.groupby('name_instructor').sum().sort_values(by='num_students',ascending=False)

df_sorted_num_student.iloc[50:100]

df_instructor_sorted=df_sorted_num_student.iloc[:200].groupby('name_instructor').count().sort_values(by='title').iloc[-6:]
patches, texts, autotexts = plt.pie(
    df_instructor_sorted['title'],
    explode=(0,1,0,0,0,0),
    labels=df_instructor_sorted.index,
    autopct="%1.1f%%",
    shadow=True,
    startangle=90
)

for t in texts:
    t.set_fontsize(12)
    t.set_color('black')
    
for t in autotexts:
    t.set_fontsize(13)
    t.set_color('white')
plt.title("인기 강좌 200위 내 인기 강사 6명")
plt.show()
df_sorted_num_student.groupby('name_instructor').sum().sort_values(by='num_students',ascending=False).iloc[:10]
```
